```python
import pandas as pd
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
sp500 = yf.download('^GSPC', start='2010-01-01', end='2025-01-01')
vix = yf.download('^VIX', start='2010-01-01', end='2025-01-01')
```

```python

sp500 = sp500.rename(columns={'^GSPC': 'SP500'})
sp500 = sp500.drop(['High', 'Low', 'Volume'], axis=1, level=0)

vix = vix.rename(columns={'^VIX': 'VIX'})
vix = vix.drop(['High', 'Low', 'Volume'], axis=1, level=0)
```

```python
sp500.columns = [f"SP500_{col[0]}" for col in sp500.columns]
vix.columns = [f"VIX_{col[0]}" for col in vix.columns]
```

```python
df = pd.concat([sp500, vix], axis=1).dropna()
```
