```python
import pandas as pd
import yfinance as yf
import numpy as np

sp500 = yf.download('^GSPC', start='2010-01-01', end='2025-01-01')
vix = yf.download('^VIX', start='2010-01-01', end='2025-01-01')

df = pd.concat([sp500, vix], axis=1)
```

```python
df.head()
```

```python
df = df.rename(columns=({'^GSPC': 'SP500','^VIX': 'VIX'}))
#df = df.drop(['High', 'Low', 'Volume'], axis=1, level=0)
df.columns = [f"{col[1]}_{col[0]}" for col in df.columns]

df.head()
```

```python
# Calculate returns
close = df['SP500_Close']

df['daily_return'] = close.pct_change()
df['log_daily_return'] = np.log(close /close.shift(1))
df['5d_return'] = close.pct_change(5)
df['20d_return'] = close.pct_change(20)

df.head(30)
```

```python
#calulate volatility
dpy = 252  # trading days per year
df['5d_volatility'] = df['log_daily_return'].rolling(window=5).std()
df['20d_volatility'] = df['log_daily_return'].rolling(window=20).std()
df['60d_volatility'] = df['log_daily_return'].rolling(window=60).std()
df['vol_of_vol_20d'] = df['20d_volatility'].rolling(window=20).std()

df['realized_quarticity_20d'] = df['log_daily_return'].rolling(20).apply(lambda x: np.mean(x**4))
df['realized_skew_20d'] = df['log_daily_return'].rolling(20).skew()
df['realized_kurtosis_20d'] = df['log_daily_return'].rolling(20).kurt()

ret_mean_20d = df['log_daily_return'].rolling(20).mean()
df['return_Zscore_20d'] = (df['log_daily_return'] - ret_mean_20d) / df['20d_volatility']

df.head(30)
```

```python
# Moving averages and trend

df['MA_20'] = close.rolling(window=20).mean()
df['MA_50'] = close.rolling(window=50).mean()
df['MA_200'] = close.rolling(window=200).mean()

def rolling_slope(series, window):
    return series.rolling(window).apply(lambda x: np.polyfit(range(window), x, 1)[0], raw=False)

df['slope_MA_20'] = rolling_slope(df['MA_20'], 20)
df['slope_MA_50'] = rolling_slope(df['MA_50'], 50)
df['slope_MA_200'] = rolling_slope(df['MA_200'], 200)

df['price_MA_20'] = close/df['MA_20']
df['price_MA_50'] = close/df['MA_50']
df['price_MA_200'] = close/df['MA_200']

df[['slope_MA_20','slope_MA_50']][100:150]
```

```python
# Drawdown and autocorrelation

df['rolling_max'] = close.rolling(252).max()   # 1-year rolling peak
df['current_drawdown'] = (close - df['rolling_max']) / df['rolling_max']
df['max_drawdown_252d'] = df['current_drawdown'].rolling(252).min()

MAR_annual = 0.042  # Minimum Acceptable Return (T-Bill)
MAR_daily = (1 + MAR_annual)**(1/dpy) - 1
df['downside_dev_20d'] = np.sqrt(((np.minimum(df['daily_return'] - MAR_daily, 0))**2).rolling(window=20).mean())

df['autocorr_1d'] = df['log_daily_return'].rolling(50).apply(lambda x: x.autocorr(1))
df['autocorr_5d'] = df['log_daily_return'].rolling(100).apply(lambda x: x.autocorr(5))
df['autocorr_20d'] = df['log_daily_return'].rolling(200).apply(lambda x: x.autocorr(20))
# almost no autocorrelation after 1 day

# we may wanna try to recalculate autocorrelation after splitting data per regime
```

```python
# VIX related features
df['dVIX_1d'] = np.log(df['VIX_Close']).diff()
df['dVIX_5d'] = df['dVIX_1d'].rolling(window=5).sum()

df['VIX_Zscore_20d'] = (df['dVIX_1d'] - df['dVIX_1d'].rolling(window=20).mean())/df['dVIX_1d'].rolling(window=20).std()
df['VIX_vol_20d'] = df['dVIX_1d'].rolling(20).std()

# Volatility Risk Premium Proxy (VRP = VIX^2 - Realized Volatility^2)
df['realized_vol_20d_ann'] = df['20d_volatility'] * np.sqrt(dpy)
df['VRP_Proxy'] = (df['VIX_Close']/100)**2 - df['realized_vol_20d_ann']**2 # VIX is in percentage points

df[['VIX_Close','dVIX_1d']]
```

```python
# Interactions

df['corr_ret_dVIX_20d'] = df['log_daily_return'].rolling(window=20).corr(df['dVIX_1d']) #not sure of rolling window size for vix (i tried 5d)
df['ret_x_VIX'] = df['log_daily_return'] * df['dVIX_1d']
df['vol_x_VIX'] = df['20d_volatility'] * df['dVIX_1d']
df['vol_ratio'] = df['realized_vol_20d_ann'] / (df['VIX_Close']/100)
df['VIX_Zscore_X_DD'] = df['VIX_Zscore_20d'] * df['current_drawdown']

```

```python
#rsi indicator
delta = close.diff()
gain = (delta.where(delta > 0, 0)).rolling(window=20).mean()
loss = (-delta.where(delta < 0, 0)).rolling(window=20).mean()
rs = gain / loss
df['RSI_20'] = 100 - (100 / (1 + rs))
```

# use of adf and kpss tests to create stationarity features
from statsmodels.tsa.stattools import adfuller, kpss
def adf_test(series):
    result = adfuller(series.dropna())
    return result[1]  # return p-value
def kpss_test(series):
    result = kpss(series.dropna(), nlags="auto")
    return result[1]  # return p-value
df['adf_log_return_pvalue'] = df['log_daily_return'].rolling(window=252).apply(adf_test, raw=False)
df['kpss_log_return_pvalue'] = df['log_daily_return'].rolling(window=252).apply(kpss_test, raw=False)

# RESULT: enough stationarity to allow for the use of the hurst exponent


```python
#create hurst exponent feature
from hurst import compute_Hc

def hurst_exponent(series, window):
    def hurst_calc(x):
        H, c, data = compute_Hc(x, kind='change', simplified=True)
        return H
    return series.rolling(window=window).apply(hurst_calc, raw=False)

df['hurst_100d'] = hurst_exponent(df['log_daily_return'], 100)

# create ADX feature
def calculate_adx(data, window=14):
    high = data['SP500_High']
    low = data['SP500_Low']
    close = data['SP500_Close']
    plus_dm = high.diff().where((high.diff() > low.diff()) & (high.diff() > 0), 0)
    minus_dm = low.diff().where((low.diff() > high.diff()) & (low.diff() > 0), 0)
    tr1 = high - low
    tr2 = (high - close.shift()).abs()
    tr3 = (low - close.shift()).abs()
    tr = pd.concat([tr1, tr2, tr3], axis=1).max(axis=1)
    atr = tr.rolling(window=window).mean()
    plus_di = 100 * (plus_dm.rolling(window=window).mean() / atr)
    minus_di = 100 * (minus_dm.rolling(window=window).mean() / atr)
    dx = (abs(plus_di - minus_di) / (plus_di + minus_di)) * 100
    adx = dx.rolling(window=window).mean()
    return adx
```

```python
df['ADX_14'] = calculate_adx(df, window=14)
```
